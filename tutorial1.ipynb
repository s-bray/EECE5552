```python
import mujoco
import mujoco.viewer

import numpy as np
```

```python
# Load the combined scene (it includes go1_position.xml automatically)
model = mujoco.MjModel.from_xml_path("scene_torque.xml")
data = mujoco.MjData(model)

mujoco.mj_forward(model, data)
```

```python
names = [model.jnt(i).name for i in range(model.njnt)]
print(names)
# print(model.njnt)
```

```python
jntID = data.jnt('FL_hip_joint').id

val = 0.0
itr = 0.01
with mujoco.viewer.launch_passive(model, data) as viewer:
    while viewer.is_running():
        data.ctrl[jntID] = np.array(val)

        mujoco.mj_step(model, data)

        if val > 0.5:
            itr = -0.01
        elif val < -0.5:
            itr = 0.01
        
        val += itr
        
        viewer.sync()
```
